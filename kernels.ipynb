```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import LinearSVC, SVC
from sklearn.preprocessing import StandardScaler
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv(r'F:\Documenti\machine_learning_avanzato\data2.csv', names=['x', 'y', 'label'])
df['label'][df['label']== 0] = -1
```

```python
for index, row in df.iterrows():
    color = ['r', 'g', 'b', 'y', 'm', 'k', 'c']
    plt.scatter(row.x, row.y, c=color[int(row['label'])])
```

```python
X = df[['x', 'y']].to_numpy()
y = df['label'].astype('int32').to_numpy()
```

```python
X_sc = StandardScaler().fit_transform(X, y)
```

```python
svc = LinearSVC(C=1)
```

```python
svc.fit(X_sc, y)
```

```python
svc.classes_
```

```python
plot_decision_regions(X=X_sc, y=y, clf=svc)
```

```python
svc_rbf = SVC(kernel='rbf', gamma=1)
```

```python
svc_rbf.fit(X_sc, y)
```

```python
plot_decision_regions(X=X_sc, y=y, clf=svc_rbf)
```
